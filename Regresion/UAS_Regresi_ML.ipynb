```python
import pandas as pd

file_path = '/content/drive/MyDrive/UAS ML/midterm-regresi-dataset.csv'
df = pd.read_csv(file_path)
display(df.head())
```

```python
y = df.iloc[:, 0]
X = df.iloc[:, 1:]
```

```python
df.isnull().sum().sum()
```

```python
df.fillna(df.median(), inplace=True)
```

```python
import matplotlib.pyplot as plt

plt.hist(y, bins=20)
plt.xlabel("Year")
plt.ylabel("Count")
plt.title("Distribution of Song Release Year")
plt.show()
```

```python
plt.boxplot(y)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.2,
    random_state=42
)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential([
    Dense(128, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(64, activation='relu'),
    Dense(1)
])

model.compile(
    optimizer='adam',
    loss='mse',
    metrics=['mae']
)
```

```python
history = model.fit(
    X_train,
    y_train,
    validation_split=0.2,
    epochs=30,
    batch_size=64
)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("MSE :", mse)
print("RMSE:", rmse)
print("MAE :", mae)
print("R2  :", r2)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
plt.scatter(y_test, y_pred, alpha=0.3)
plt.xlabel("Actual Year")
plt.ylabel("Predicted Year")
plt.title("Actual vs Predicted")
plt.show()
```

```python
import optuna
import tensorflow as tf
from sklearn.metrics import mean_squared_error
import numpy as np

def objective(trial):

    units1 = trial.suggest_int("units1", 64, 256)
    units2 = trial.suggest_int("units2", 32, 128)

    model = tf.keras.Sequential([
        tf.keras.layers.Dense(units1, activation='relu', input_shape=(X_train.shape[1],)),
        tf.keras.layers.Dense(units2, activation='relu'),
        tf.keras.layers.Dense(1)
    ])

    model.compile(
        optimizer='adam',
        loss='mse'
    )

    model.fit(
        X_train,
        y_train,
        epochs=10,
        batch_size=64,
        verbose=0
    )

    y_pred = model.predict(X_test, verbose=0)

    rmse = np.sqrt(mean_squared_error(y_test, y_pred))

    return rmse
```

```python
study = optuna.create_study(direction="minimize")
study.optimize(objective, n_trials=10)

print(study.best_params)
print(study.best_value)
```

```python
import tensorflow as tf

final_model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(X_train.shape[1],)),
    tf.keras.layers.Dense(238, activation='relu'),
    tf.keras.layers.Dense(43, activation='relu'),
    tf.keras.layers.Dense(1)
])

final_model.compile(
    optimizer='adam',
    loss='mse',
    metrics=['mae']
)

history = final_model.fit(
    X_train,
    y_train,
    validation_split=0.2,
    epochs=30,
    batch_size=64
)
```

```python
y_pred_final = final_model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

mse = mean_squared_error(y_test, y_pred_final)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred_final)
r2 = r2_score(y_test, y_pred_final)

print("MSE :", mse)
print("RMSE:", rmse)
print("MAE :", mae)
print("R2  :", r2)
```
